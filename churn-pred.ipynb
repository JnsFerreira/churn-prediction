```python
def check_fill(df):
    #Cria um dicionario vazio
    fill_cols = {}
    
    #Percorre todas as colunas do dataframe
    for column in df.columns:
        fill_cols.update({column :df[column].isna().count() / df.shape[0] *100})
    
    #Converte o dicionário em um pandas DataFrame e renomeia as colunas
    df_fill = pd.DataFrame.from_dict(fill_cols, orient='index')
    df_fill.reset_index(inplace=True)
    df_fill.rename(columns={'index':'columnName',0:'fillPercent'}, inplace=True)
    
    #Retorna o DataFrame formatado
    return df_fill
```

```python
#Aplicando a função 'check_fill'
df_fill = check_fill(df)

#Checando os valores obtidos
df_fill
```
